### 卷积和反卷积中padding 为same/valid的一些计算：


```python
import tensorflow as tf
def test(inputs,size=64,k=5,strides=2,padding='same'):
    x1 = tf.layers.conv2d(inputs,size,k,strides=strides,padding=padding)
    print(x1.get_shape().as_list())
    return x1
def test_d(inputs,size,k=5,s=2,padding='same'):
    x1 = tf.layers.conv2d_transpose(inputs,size,k,strides=s,padding=padding)
    print(x1.get_shape().as_list())
    return x1
```

```python
image = tf.placeholder(tf.float32, [None, 28, 28, 3])
test(image,64,5,2)
```

```python
image = tf.placeholder(tf.float32,[None,7,7,32])
test(image,64,5,2)
```

```python
image = tf.placeholder(tf.float32,[None,7,7,32])
image2 = tf.placeholder(tf.float32,[None,7,7,32])
test(image,64,4,1,'valid')
test(image2,64,4,1,'same')
```

```python
image = tf.placeholder(tf.float32,[None,4,4,32])
test_d(image,64,4,1,'valid')
image = tf.placeholder(tf.float32,[None,4,4,32])
test_d(image,64,5,2,'same')
```

```python
image = tf.placeholder(tf.float32,[None,28,28,3])
x1 = test(image,64)
x2 = test(x1,256)
x3 = test(x2,512)
```
